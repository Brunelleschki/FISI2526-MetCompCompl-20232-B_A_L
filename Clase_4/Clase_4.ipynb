```python
from funciones import posicion
import matplotlib.pyplot as plt
import numpy as np
```

```python
A = 2
k = 800
m = 2
T = 2*np.pi*np.sqrt(m/k)
w = np.sqrt(k/m)
a = np.linspace(-T, T, 100)
```

```python
x = posicion(2, 800, 2, a)
plt.plot(a,x)
```

```python
def X(a):
    return posicion(2, 800, 2, a)
```

```python
def DerivadaDerecha(f,x,h):
    d = 0.
    if h != 0:
        d = (f(x+h) - f(x))/h  
    return d
```

```python
def DerivadaIzquierda(f,x,h):
    d = 0.
    if h != 0:
        d = (f(x) - f(x-h))/h
    return d
```

```python
def DerivadaCentral(f,x,h):
    d = 0.
    if h != 0:
        d = (f(x+h) - f(x-h))/(2*h)
    return d
```

```python
def DerivadaExacta(x):
    return A*w*np.cos(w*x)
```

```python
DD = DerivadaDerecha(X, a, 10e-4)
DI = DerivadaIzquierda(X, a, 10e-4)
DC = DerivadaCentral(X, a, 10e-4)
DE = DerivadaExacta(a)
```

```python
plt.scatter(a, DD, label = 'DerivadaDerecha', c = "pink", s = 1)
plt.scatter(a, DI, label = 'DerivadaIzquierda', c = "blue", s = 1)
plt.scatter(a, DC, label = 'DerivadaCentral', c = "green", s = 1)
plt.scatter(a, DE, label = 'DerivadaExacta', c = 'red', s = 1)
plt.legend()
```
